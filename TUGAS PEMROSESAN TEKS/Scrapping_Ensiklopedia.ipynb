```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://esi.kemdikbud.go.id/wiki/Category:Peristiwa')
bs = BeautifulSoup(page.text, 'html.parser')
bs
```

```python
bs.find_all('a', title=True)
```

```python
list_items_with_title = bs.find_all('a', title=True)
for li in list_items_with_title:
    print(li['title'])
```

```python
unique_titles = []

for li in list_items_with_title:
    title = li['title']

    if title != 'Main Page' and title not in unique_titles:
        unique_titles.append(title)

sorted_titles = sorted(unique_titles)
```

```python
print('Berikut adalah daftar peristiwa-peristiwa penting dalam sejarah Indonesia:\n')

for i, title in enumerate(sorted_titles, start=1):
    print(f"{i}. {title}")
```
